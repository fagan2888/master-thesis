```python
import numpy as np
import pandas as pd
```

```python
df_monthly = pd.read_excel('macro.xls', sheetname = 'Monthly', index_col = 'DATE')
df_quarterly = pd.read_excel('macro.xls', sheetname = 'Quarterly', index_col = 'DATE')
```

```python
# add monthly data to the quarterly df
df_quarterly = df_quarterly.join(df_monthly)

# drop rows that have nulls (drop the beginning until all time series have started)
df_quarterly.dropna(inplace=True)
df_monthly.dropna(inplace=True)

# only keep data between 1959-01 and m:2014-11, q:2014-09 to have the data match the factors
# (remove the last period because there are a lot of nulls in it)
df_quarterly = df_quarterly['1959-01-01': '2014-09-01']
df_monthly = df_monthly['1959-01-01': '2014-11-01']
```
